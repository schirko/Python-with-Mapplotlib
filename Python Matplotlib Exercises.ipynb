### Basic Plots
#### Line Plot
Thsi first example is the most basic line plot. It plots the year and world population. plt.xlabel and plt.ylabel are used to annotate the axis, and plot.title to give the chart a title. plt.yticks is used to set intervals on the y axis.

```python
# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

year= [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
plt.plot(year, pop)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Projections')
plt.yticks([0, 2, 4, 6, 8, 10])
plt.show()
```

#### Scatter Plot
For the scatter plot all we do is change plt.plot for plt.scatter.

```python
# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

year= [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
plt.scatter(year, pop)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Projections')
plt.yticks([0, 2, 4, 6, 8, 10])
plt.show()
```

We implement more customizations by adding grid lines, ticks, and population scale.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data into pandas dataframe
df_le = pd.read_csv('Life_Expectancy_WHO_Data.csv')
df_2010 = df_le[df_le['Year'] == 2010].dropna()
np_pop = np.array(df_2010.Population.dropna())

plt.scatter(df_2010.GDP, df_2010.Life_Expectancy, alpha = .8, s = np_pop / 500000)
plt.xscale('log') 
plt.title('Life Expectancy vs. GDP per capita in 2010')
plt.xlabel('GDP per capita')
plt.ylabel('Life Expectancy')
plt.xticks([1000, 10000, 100000], ['1k','10k', '100k'])

# Additional customizations
#plt.text(1345.77153, 66.4, 'India')
#plt.text(456.512487, 75, 'China')

# Add grid() call
plt.grid(True)

plt.show()
```

#### Histogram

A histogram is a graph showing frequency distributions. It is a graph showing the number of observations within each given interval.

In Matplotlib, we use the hist() function to create histograms. In the example below numpy is also used to generate random numbers for us to plot.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show() 
```

We can tweak the the histogram to show a speficic number of bins, in this case we choose 5.  By default, Python chooses 10 bins.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x, bins=5)
plt.show() 
```
